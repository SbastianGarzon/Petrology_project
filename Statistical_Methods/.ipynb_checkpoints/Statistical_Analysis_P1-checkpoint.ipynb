# Statistical Analysis of Geochemical Data

Statistical data analysis is an important tool in science that improves the way to organize and interpret measurements. The present notebook shows the importance of this topic in geochemistry, and how to analyze geochemical data taking into account generic issues of statistics.

## Averages: Robust Estimates

The averages are widely used in geochemistry. For example, the averages values of elements in  chondritic meteorites or the presentation of results in an investigation.  
  
The median, mode, arithmetic mean and geometric mean are the main ways to calculate averages. In geochemistry is frequently used the arithmetic and geometric mean. However, these methods are statisticaly inapropiate, because geochemical data are non-normally distributed or include outliers. Therefore, the introduction to Robust Estimates in geochemical data analysis is very important.
